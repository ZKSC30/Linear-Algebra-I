```python
''' Collatz Conjecture '''


n = 775 # 22 -> 11 -> 34 -> 17 -> 52 -> 26 -> 13 -> 40 -> 20 -> 10 -> 5 -> 16 -> 8 -> 4 -> 2 -> 1

''' 
    If n is divisible by 2 -> divide by 2
    Else -> Multiply by 3 and add 1
    
    Collatz Conjecture: This procedure always reaches 1 no matter the starting point

'''

steps = 0
while n != 1:    # Make sure you have a way to exit!
    steps += 1   # steps = steps + 1
    print(steps,n)
    if n % 2 == 0:
        n = int(n/2)
    else:
        n = 3*n+1
```

```python
''' 
    Factorial Function 
    
    1! = 1 
    2! = 2.1
    3! = 3.2.1
    .
    .
    .

'''

n   = 12
fac = 1
for i in range(1,n+1):
    fac *= i   # fac = fac * i
```
