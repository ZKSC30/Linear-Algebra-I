```python
""" 3D Example """
u = np.array([1,1,1])
v = np.array([1,0,0])
w = np.array([0,0,1])

up = u
```

```python
vp = v - ((v@up)/(up@up))*up
print(vp) # (2/3,-1/3,-1/3)
```

```python
wp = w - ((w@up)/(up@up))*up - ((w@vp)/(vp@vp))*vp
print(wp) # (0,-1/2,1/2)
```

```python
def plot_2d_vector(vec,orig=np.array([0,0])):
    """ 
        This orig=...  is a default argument 
        
    """
    plt.plot([orig[0],vec[0]],[orig[1],vec[1]])
```

```python
vx = np.array([1,2])
vy = np.array([3,2])

plot_2d_vector(vx)
plot_2d_vector(vy,vx)
plot_2d_vector(vy)
```
