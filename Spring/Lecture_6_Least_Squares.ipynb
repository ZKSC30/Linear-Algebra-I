# Problem #1:

## Given a basis {v1,v2,...,vn}

### Determine if the basis is orthogonal
### If not make orthogonal

## Given some extra vector w

### Determine if the vector belongs to the space generated by the given basis
### Find the coefficients of the closest element in the space to w

```python
def problem_1_part_1(vector_list):
    """ 
        vector_list : np.array( np.array(float) ) list of vectors (=lists of numbers)
    
    """
    
    """ First if they are otrhogonal """
    are_all_orthogonal = True
    #for V1 in vector_list: # for FREE_NAME in LIST_NAME:
    #    print(V1)
    
    # len(ANY_LIST) -> length of the list
    # range(ANY_NUM) -> (0,1,2,...,ANY_NUM-1)
    
    for i in range(len(vector_list)): # for FREE_NAME 
        for j in range(len(vector_list)):
            if i >= j: # reads: if i is equal to j
                continue # skips all the code and moves on to the next j value
            print("(",i,"",j,")",vector_list[i] @ vector_list[j])
            print("are_all_orthogonal before:",are_all_orthogonal)
            are_all_orthogonal = are_all_orthogonal & ((vector_list[i] @ vector_list[j]) == 0)
            print("are_all_orthogonal after:",are_all_orthogonal)
    
    # Boolean Algebra: Answers to yes/no questions
    # True, False 
    
    # True  and False -> False
    # True  and True  -> True
    # False and False -> False
    
    # Ex: are_all_orthogonal = True
    # ip = 32
    # are_all_orthogonal = True and (32==0) = True and False = False
    return are_all_orthogonal
```

```python
v1 = np.array([1,2,3])
v2 = np.array([3,0,-1])
v3 = np.array([7,8,9])

candidate_basis = np.array([v1,v2,v3])
```

```python
problem_1_part_1(candidate_basis)
```
